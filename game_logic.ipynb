```python
import requests
import json
```

```python
# create a counter for the streak
streak = 0

#create dictionary for the id of catergories
trivia_categories = {
    "General Knowledge": 9,
    "Entertainment: Books": 10,
    "Entertainment: Film": 11,
    "Entertainment: Music": 12,
    "Entertainment: Musicals & Theatres": 13,
    "Entertainment: Television": 14,
    "Entertainment: Video Games": 15,
    "Entertainment: Board Games": 16,
    "Science & Nature": 17,
    "Science: Computers": 18,
    "Science: Mathematics": 19,
    "Mythology": 20,
    "Sports": 21,
    "Geography": 22,
    "History": 23,
    "Politics": 24,
    "Art": 25,
    "Celebrities": 26,
    "Animals": 27,
    "Vehicles": 28,
    "Entertainment: Comics": 29,
    "Science: Gadgets": 30,
    "Entertainment: Japanese Anime & Manga": 31,
    "Entertainment: Cartoon & Animations": 32
}
```

```python
#make function to call a api call to the trivia database for a specific catgeory and number of questions
def load_pack(category, num_q):
    url = "https://opentdb.com/api.php?amount=" + str(num_q) + "&category=" + str(category)
    response = requests.get(url)
    data = json.loads(response.text)
    questions = data["results"]
    return questions

# create a queue of questions so that there areent any repeats
def create_queue(pack):
    questions = pack
    queue = []
    for question in questions:
        queue.append(question)
    return queue

#make a function to get the next question in the queue
def get_next_question(queue):
    question = queue.pop(0)
    return question

#make a function to check if the answer is correct
def check_answer(question, answer):
    correct_answer = question["correct_answer"]
    if answer == correct_answer:
        return True
    else:
        return False

# make a function to update the streak, display a message if a streak of 5 is reached, and then keep displaying the streak in every intervals of 5 after that
def update_streak(streak):
    streak += 1
    if streak % 5 == 0:
        #print a message tha tuses the fire emoji to show that the user is on a streak, use the utf-8 encoding for the fire emoji
        print("You are on a streak of " + str(streak) + "ðŸ”¥")
    return streak
```

```python
#create a loop to keep the game going, first display a list of categories and then ask the user to pick a category, then ask the user to pick the number of questions they want to answer, then load the pack of questions, then create a queue of questions, then keep asking the user questions until they get one wrong, then display the streak and ask if they want to play again
while True:
    print("Pick a category:")
    for category in trivia_categories:
        print(str(trivia_categories[category]) + ". " + category)
    user_category = int(input("Category: "))
    category_id = user_category
    num_q = int(input("How many questions do you want to answer? "))
    pack = load_pack(category_id, num_q)
    queue = create_queue(pack)

    #make sure the game keeps goign even if the user gets a question wrong, and then ask if they want to play again
    while len(queue) > 0:
        question = get_next_question(queue)
        print(question["question"])

        #randomize the order of the answers, and make sure to display as A through D
        answers = question["incorrect_answers"]
        answers.append(question["correct_answer"])
        answers.sort()
        for i in range(len(answers)):
            print(chr(65 + i) + ". " + answers[i])

        
        user_answer = input("Answer: ")
        if check_answer(question, user_answer):
            streak = update_streak(streak)
        else:
            print("You got it wrong! Your streak was " + str(streak))
            streak = 0
            
    play_again = input("Do you want to play again? (yes or no) ")
    if play_again == "no":
        break
    else:
        continue

print("Thanks for playing!")
```
